```python
import cv2 as cv
import numpy as np
from matplotlib import pyplot
from matplotlib import image

# my module
import threshold
import morph_v2 as morph
import thinning
```

```python
def showAndSaveImg(img, saveName = 0, isSave = 0):
    pyplot.imshow(img, cmap='gray')
    pyplot.axis('off')
    if (isSave) == 1:
        pyplot.savefig(saveName, bbox_inches='tight')
    pyplot.show()
```

```python
lion_rgb = cv.imread('./Inputs/hope.png')
showAndSaveImg(lion_rgb, 0, 0)
lion_rgb.shape
```

```python
lion_gray = cv.cvtColor(lion_rgb, cv.COLOR_RGB2GRAY)
showAndSaveImg(lion_gray, './Images/gray.png', 0)
```

```python
lion_binary = threshold.basicThreshold(lion_gray, 140)
showAndSaveImg(lion_binary, 0, 0)
```
