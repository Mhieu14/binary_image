```python
import cv2 as cv
import numpy as np
from matplotlib import pyplot
from matplotlib import image

# my module
import threshold
import morph_v2 as morph
```

```python
def showAndSaveImg(img, saveName, isSave = 0):
    pyplot.imshow(img, cmap='gray')
    pyplot.axis('off')
    if (isSave) == 1:
        pyplot.savefig(saveName, bbox_inches='tight')
    pyplot.show()
```

```python
lion_rgb = cv.imread('./Inputs/pagoda.png')
showAndSaveImg(lion_rgb, 0, 0)
lion_rgb.shape
```

```python
lion_gray = cv.cvtColor(lion_rgb, cv.COLOR_RGB2GRAY)
showAndSaveImg(lion_gray, './Images/gray.png')
```

```python
gauss_th_cv = cv.adaptiveThreshold(lion_gray, 
                            maxValue=255, 
                            adaptiveMethod=cv.ADAPTIVE_THRESH_GAUSSIAN_C, 
                            thresholdType=cv.THRESH_BINARY,
                            blockSize=15,
                            C=8)
print("Gaussian:")
showAndSaveImg(gauss_th_cv, 0, 0)

mean_th_cv = cv.adaptiveThreshold(lion_gray, 
                            maxValue=255, 
                            adaptiveMethod=cv.ADAPTIVE_THRESH_MEAN_C, 
                            thresholdType=cv.THRESH_BINARY,
                            blockSize=15,
                            C=8)
print("Mean:")
showAndSaveImg(mean_th_cv, 0, 0)
```
