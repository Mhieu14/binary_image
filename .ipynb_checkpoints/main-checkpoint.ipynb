```python
import cv2 as cv
import numpy as np
from matplotlib import pyplot
from matplotlib import image

# my module
import threshold
import morph
```

```python
def showAndSaveImg(img, saveName, isSave = 1):
    pyplot.imshow(img, cmap='gray')
    pyplot.axis('off')
    if (isSave) == 1:
        pyplot.savefig(saveName, bbox_inches='tight')
    pyplot.show()
```

```python
lion_rgb = cv.imread('./Images/lion.png')
showAndSaveImg(lion_rgb, 0, 0)
lion_rgb.shape
```

```python
lion_gray = cv.cvtColor(lion_rgb, cv.COLOR_RGB2GRAY)
showAndSaveImg(lion_gray, './Images/gray.png')
```

```python
lion_gray
```

```python
basic_th = {}
for i in [50, 80, 110, 140, 170, 200]:
    name = './Images/basic_th_' + str(i) + '.png'
    basic_th_i = threshold.basicThreshold(lion_gray, i)
    basic_th[i] = basic_th_i
    print(name)
    showAndSaveImg(basic_th_i, name)
```

```python
for size in [5, 15, 25]:
    for c in [4, 8, 15]: 
        name = './Images/mean_th_' + str(size) + '_' + str(c) + '.png'
        mean_th = threshold.meanAdaptiveThreshold(lion_gray, size, c)
        print(name)
        showAndSaveImg(mean_th, name)
```

```python
gauss_th_cv = cv.adaptiveThreshold(lion_gray, 
                            maxValue=255, 
                            adaptiveMethod=cv.ADAPTIVE_THRESH_GAUSSIAN_C, 
                            thresholdType=cv.THRESH_BINARY,
                            blockSize=15,
                            C=8)
print("Gaussian:")
showAndSaveImg(gauss_th_cv, 0, 0)

mean_th_cv = cv.adaptiveThreshold(lion_gray, 
                            maxValue=255, 
                            adaptiveMethod=cv.ADAPTIVE_THRESH_MEAN_C, 
                            thresholdType=cv.THRESH_BINARY,
                            blockSize=15,
                            C=8)
print("Mean:")
showAndSaveImg(mean_th_cv, 0, 0)
```

```python
kernel = np.ones((3,3),np.uint8)
kernel
```

```python
basic_th[140]
```

```python
dilation = morph.morph_filter(
    operator='di',
    img=basic_th[140],
    sel=kernel,
    n_iterations=1)
showAndSaveImg(dilation, './Images/dilation')
```

```python
erosion = morph.morph_filter(
    operator='er',
    img=basic_th[140],
    sel=kernel,
    n_iterations=1)
showAndSaveImg(erosion, './Images/erosion')
```

```python
opening = {}
for i in [3, 5, 7]:
    kernel = np.ones((i,i),np.uint8)
    opening_i = morph.morph_filter(
        operator='op',
        img=basic_th[140],
        sel=kernel,
        n_iterations=1)
    name = './Images/opening_' + str(i)
    opening[i] = opening_i
    print(name)
    showAndSaveImg(opening_i, name)
```

```python
closing = {}
for i in [3, 5, 7]:
    kernel = np.ones((i,i),np.uint8)
    closing_i= morph.morph_filter(
        operator='cl',
        img=basic_th[140],
        sel=kernel,
        n_iterations=1)
    name = './Images/closing_' + str(i)
    closing[i] = closing_i
    print(name)
    showAndSaveImg(closing_i, name)
```

```python
gradient = morph.morph_filter(
        operator='mg',
        img=basic_th[140],
        sel=kernel,
        n_iterations=1)
showAndSaveImg(gradient, './Images/gradient')
```
